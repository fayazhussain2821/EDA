```python
column_name = 'Cuisines'
```

```python
missing_indicator = df[column_name].isnull()
```

```python
# Set a good-looking style
sns.set_theme(style='whitegrid')

plt.figure(figsize=(7, 5))
ax = sns.countplot(x=missing_indicator)

# Add the counts on top of each bar
# ax.containers holds the bar objects
ax.bar_label(ax.containers[0], fmt='%d')

plt.title(f'Missing Values in "{column_name}" Column')
plt.xlabel('Is Missing')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['False (Not Missing)', 'True (Missing)'])
plt.tight_layout()

plt.show()
```

```python
# Create a DataFrame for plotting the missing value positions
missing_positions = df[df['Cuisines'].isnull()].index
plt.figure(figsize=(10, 2))
plt.scatter(missing_positions, np.zeros(len(missing_positions)), color='red', marker='x', s=100)
plt.yticks([]) 
plt.xlabel('Row Index')
plt.title('Positions of Missing Values in "Cuisines" Column')
plt.show()
```

```python
df_country=pd.read_excel("./datasets/Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_values=final_df.Country.value_counts().values
```

```python
# Create a DataFrame for convenience
top_3_data = pd.DataFrame({
    'Country': country_names[:3],
    'Value': country_values[:3]
})

# Use seaborn for plotting with a built-in theme
sns.set_theme(style='whitegrid')
plt.figure(figsize=(10, 6))

# Create the bar plot using seaborn's barplot()
# A "palette" can be used for beautiful, sequential colors
ax=sns.barplot(x='Country', y='Value', data=top_3_data, palette='viridis')

# Add the counts on top of each bar
ax.bar_label(ax.containers[0], fmt='%d')

# Add labels and title
plt.title('Top 3 Countries by Value', fontsize=18, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Value', fontsize=14)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
city_names=final_df.City.value_counts().index
city_values=final_df.City.value_counts().values
```

```python
# Create a DataFrame for convenience
top_5_data = pd.DataFrame({
    'City': city_names[:5],
    'Value': city_values[:5]
})

# Use seaborn for plotting with a built-in theme
sns.set_theme(style='whitegrid')
plt.figure(figsize=(10, 6))

# Create the bar plot using seaborn's barplot()
# A "palette" can be used for beautiful, sequential colors
ax=sns.barplot(x='City', y='Value', data=top_5_data, palette='Set2')

# Add the counts on top of each bar
ax.bar_label(ax.containers[0], fmt='%d')

# Add labels and title
plt.title('Top 5 Cities by Value', fontsize=18, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Value', fontsize=14)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
